```python
mnist = load_dataset("mnist", max_sample=1000)
yale = load_dataset("yale")
orl = load_dataset("orl")
```

```python
def apply_SDRC(dataset, epochs=100):
    X, y, shp = dataset
    
    alg = SDRC(K=10, l=0.001)
    Z, B, z = alg.fit(X, epochs=epochs)
    
    alg.plot_results(B, y)
    print_results(X, img_shape=shp, encoded_imgs=B)
    res = alg.clustering_report(z, y)

    
    return HTML(f'<h3>Accuracy : {res["accuracy"]}<h3/><h3>NMI : {res["nmi"]}<h3/><h3>ARI : {res["ari"]}<h3/>')
```

```python
apply_SDRC(orl)
```
